## Hypothesis Tests

### Slope for energy

Ho: slope = 0

Ha: slope != 0

```python
crit_val = norm.ppf(0.995)
print("Critical value:", crit_val)
beta = -0.3991
std_err= 2.716

lower = beta - crit_val*std_err
upper = beta + crit_val*std_err

print("99% Confidence interval: (", lower, ",", upper, ")")
```

Hence since 0 lies in interval, not sufficient evidence to reject the null hypothesis.

### Slope for loudness

```python
crit_val = norm.ppf(0.995)
print("Critical value:", crit_val)
beta = 0.2413
std_err= 0.145

lower = beta - crit_val*std_err
upper = beta + crit_val*std_err

print("99% Confidence interval: (", lower, ",", upper, ")")
```

## ROC Curves

```python
def plot_roc(fpr, tpr, auc, lw=2):
    plt.plot(fpr, tpr, color='darkorange', lw=lw,
             label='ROC curve (area = '+str(round(auc,3))+')')
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def create_roc_curve(training_model,df_test):
    phat_test = training_model.predict(exog=dict(df_test))
    df_test['phat_test'] = phat_test
    fpr_mod_train, tpr_mod_train, score = roc_curve(y_true=df_test['Include'], y_score=df_test['phat_test'])
    auc_mod_train = roc_auc_score(y_true=df_test['Include'], y_score=df_test['phat_test'])
    plot_roc(fpr_mod_train, tpr_mod_train, auc_mod_train)
    
    
```

## Using SKL

### Creating features matrix and target array

```python
x_train = df_train.drop(columns = ['type','id','uri','Include'])
x_train.head()
```

```python
x_test = df_test.drop(columns = ['type','id','uri','Include'])
x_test.head()
```

```python
x_train['mode'] = x_train['mode'].astype('category',copy=False)
x_test['mode'] = x_test['mode'].astype('category',copy=False)

x_train['time_signature'] = x_train['time_signature'].astype('category', copy=False)
x_test['time_signature'] = x_test['time_signature'].astype('category', copy=False)

x_train = pd.get_dummies(x_train, drop_first = True)
x_test = pd.get_dummies(x_test, drop_first = True)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train = df_train['Include']
y_train.head()
```

```python
y_test = df_test['Include']
y_test.head()
```

### Regularized and Non-Regularized Models

```python
clf0 = LogisticRegression(penalty = 'none', solver = 'newton-cg', max_iter = 10000)
clf0.fit(x_train,y_train)
```

```python
clf0.coef_
```

```python
clf1 = LogisticRegression(penalty = 'l1', solver = 'liblinear', max_iter=10000, C = 1/3)
clf1.fit(x_train,y_train)
```

```python
clf2 = LogisticRegression(penalty='l2', solver='liblinear', max_iter=10000, C = 1/3)
clf2.fit(x_train, y_train)
```

```python
clf3 = LogisticRegression(penalty = 'elasticnet', solver = 'saga', max_iter=10000, l1_ratio=0.7, C =1/3)
clf3.fit(x_train, y_train)
```

```python
dfcoef = pd.DataFrame(
    np.concatenate((clf0.coef_.T,
                   clf1.coef_.T,
                   clf2.coef_.T,
                   clf3.coef_.T),
                  axis = 1),
    columns = ['Non_Regularized', 'LASSO', 'Ridge_Regression', 'Elastic_Net'], index = x_train.columns)
dfcoef
```

```python
print(dfcoef.index)
print(dfcoef[dfcoef.columns[0]])
```

```python
plt.figure(figsize = (20,10))

for i in range(1,4):
    plt.plot(dfcoef.index, dfcoef[dfcoef.columns[i]])

plt.xticks(np.arange(0,13,1))
plt.xlabel("Coefficient Index")
plt.ylabel("Slope Estimate")
plt.title("Slope estimates for all 4 models")
plt.legend(['LASSO','Ridge_Regression', 'Elastic_Nets'])
    
plt.hlines(y = 0, xmin = 0, xmax = 13, linestyles='--')


plt.show()
```
